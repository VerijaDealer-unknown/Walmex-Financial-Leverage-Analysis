```python
#Import matplotlib from pyplot one of the most popular charting library
import matplotlib.pyplot as plt

#1 creating the first data frame from our insights
data={
    'Year': ['2022', '2023', '2024', '2025'],
    'ROE': [24.19, 25.91, 23.11, 21.18],
    'ROA': [11.69, 11.82, 10.90, 10.07]
}
df= pd.DataFrame(data)
print(df)

#2 Plot the lines, build the x axis and y axis
plt.plot(df['Year'], df['ROE'], label='ROE')
plt.plot(df['Year'], df['ROA'], label='ROA')

# Add basic chart labels
plt.title('Walmex Trend')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.legend()

plt.show()
```

The first trend that i can visualize, it’s in 2023 where more debt has acquired, it returned the best Return over equity and assets ni the four years, and it is slightly beyond from 2022 where it has the best net profit margin.

In first sight we can see a relationship between debt and profitability.

We also can see that instead of the increase from 2022-800B to 1025-1T. But the efficient metrics has dropped indicating the worst of the 4 years, meaning that they are increasing in sales volume but facing high costs.

Connection between ROE and D-E, we can see that instead of an debt above 1.10(safely metrics for retailers) they are having a Return of 21.8% using small leverage into huge returns.

Principal question, Do companies with more debt tend to have lower or higher profits? Higher debt directly relate to higher ROE for investors utilizing liabilities.
